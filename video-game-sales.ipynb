## Data Preparation and Cleaning

We can clean and prepare the data by using pandas librabry.



 Lets first start by importing the required libraries like pandas.

```python
import pandas as pd
```

Now lets look at the given data as a whole to figure out what type of data are we looking at exactly.

```python
vg_Df=pd.read_csv('./videogamesales/vgsales.csv')
vg_Df
```

```python
vg_Df.describe()
```

```python
vg_Df.info()
```

let us look at all the unique Platforms these game sales were made in.

```python
vg_Df.Platform.unique()
```

There looks to be an integer in the platform's column which is strange as platforms wont have just numbers in them. Lets investigate further. 

```python
vg_Df.loc[vg_Df['Platform'] == '2600']
```

It looks as if all the '2600' platform games are all atari 2600 games so its better if we rename '2600' to 'atari 2600' for better readability.

```python
vg_Df['Platform'].replace({"2600":"Atari 2600"},inplace=True)
vg_Df.Platform.unique()
```

let us now look at how many years of data this data set covers.

```python
years=vg_Df.sort_values('Year')
years
```

Here we see there are alot of games which have no year mentioned and which have made substantionly less amount of money which inturn would not cause a lot of deviation to the question we are trying to answer.So its better to drop them to make our analysis cleaner.

```python
vg_Df=vg_Df.dropna()
```

Now lets take a look at the years once again.

```python
vg_Df=vg_Df.sort_values('Year')
vg_Df
```

There looks to be very less data from the year 2020 with 2019,2018 data missing completely and only 3 values in the year 2020.Lets look at 2016 if its the same case as 2017.

```python
vg_Df.loc[vg_Df['Year'] == 2016]
```

We shall also drop the rank as the results provided by it are very arbitrary.

```python
vg_Df=vg_Df.drop(['Rank'],axis=1)
```

There is significant data for the year 2016 so it is better to delete the rows of 2020,2017 as they wont be helpful in making an accurate depiction of the sales in those years.Lets confine our analysis to years from 1980 to 2016.

```python
vg_Df.drop(vg_Df[vg_Df['Year'] >= 2017].index, inplace = True)
vg_Df
```

## Exploratory Analysis and Visualization

We shall take a closer look into the data to find more about what story the data tells us.We shall analyze the dataset for which year the video game sales were at its peak and also the top publishers as well as the top platforms of each year respectively.

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

#### Let us begin by analysing the total sales per year of each region 

```python
sales=vg_df.groupby("Year")[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum()
sales
```

As there are so many years lets look at the top 10 years which has the highest amount of sales. 

```python
sales.sort_values('Global_Sales',ascending=False).head(10)
```

As we can see there was high rise in the sales of video games in the year 2008 and there was a steady decline in sales of video games lets look at the line graph for proper visualization.

```python
vg_df.groupby(['Year'])['Global_Sales'].sum().plot()
plt.ylabel('Millions');
plt.title("Growth of video game sales");
```

As we can see there has been a significant increase in video games sales from the year 2004 and its peak around 2008.Then it leads to steady decline in video games sales till 2016


#### Now lets look at the most successful publisher among all the publishers in terms of global sales 

We can achieve this by using groupby statments as well as then sorting the values of the dataset hence produced.

```python
publisher=vg_df.groupby("Publisher")[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum()
publ=publisher.sort_values('Global_Sales',ascending=False).head(10)
publ
```

Here we can clearly see that nintendo is the absolute winner when it comes to global sales of the company.But making a visualization will help in making the point stand strong.

```python
x=publ.groupby(['Publisher'])['Global_Sales'].sum().head(10)
x.sort_values(ascending=True).plot(kind='barh');
plt.xlabel("Millions");
plt.title("Top Publishers");
```

As we had already predicted in our analysis Nintendo is leaps and bounds ahead of their predecessor.Both the sales are about 600 million sales apart 

 #### Now let us look at the most successful platform among all the platforms in terms of global sales

Lets use the same procedure as we used in the top publisher analysis to work on top platform analysis 

```python
Platform=vg_df.groupby("Platform")[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum()
plat_f=Platform.sort_values('Global_Sales',ascending=False).head(10).reset_index()
plat_f
```

The top platform in global sales paint a different picture as it looks like there is a constant cat and mouse chase between the playstation and xbox and then followed by nintendo.

```python
y=plat_f.groupby(['Platform'])['Global_Sales'].sum().head(10)
y.sort_values().plot(kind='barh')
plt.title("Top Platforms")
plt.xlabel('Millions');
```

As we can PS2 just like nintendo maintains a good lead when it comes most vidoe games sales but not as much as nintendo.But its suprising to see that Nintendo being the most succesfull publisher is not the platform which has the most platform sales even after having many platforms like WII,DS,GBA. It looks to be that playstation and microsoft have a feirce battle between as they take the first 3 sports followed by nintendo.

## Asking and Answering Questions

In this section we shall answer a few questions about the dataset and drive our analysis forward.


### Q1:Among all the genres which is the genre which produces highest volume of games? 


We can answer this by taking a look at the genre column.

```python
genre=vg_Df.Genre.value_counts().head(10)
genre
```

We can clearly see that action games is way ahead that any other genre in terms of most games produced but its better to visualize the things rather than looking at raw data.

```python
plt.figure(figsize=(25,9))
sns.barplot(x=genre.index,y=genre);
plt.title("Top Genres");
plt.xlabel("Genres");
plt.ylabel("Amount of games");
```

We can clearly see that action games are made the most.

###### We shall also answer a sub question stating which genre produces the most amount of sales.

```python
plt.figure(figsize=(25,9))
plt.pie(vg_Df.Global_Sales, labels=genre.index, autopct='%1.1f%%')
plt.title("Top Genres");
```

### Q2: What are the top 5 platforms which have the most amount of games?

We can answer this question by first taking a look at all the platforms which made games from the year 1980-2016

```python
var=vg_Df['Platform'].value_counts().to_frame()
var
```

Now lets make a bar graph of the top 5 games platform which had the most amount of games on them.

```python
plt.figure(figsize=(12,9))
vg_Df['Platform'].value_counts().head(5).plot(kind='barh');
plt.title("Top 5 platforms to host games");
```

As we can see its tough to differentiate the top 2 and also top 3,4,5 in this situation so lets reduce the intervals and see the difference.

```python
plt.figure(figsize=(14,9))
vg_Df['Platform'].value_counts().head(2).plot(kind='barh');
plt.xlim([2120, 2140]);
```

AS we can see the DS is ahead of PS2 by mere 3 games.

```python
plt.figure(figsize=(14,9));
var.iloc[2:5].plot(kind='barh');
plt.xlim([1200, 1310]);
```

We can see that PS3 is ahead of Wii by about 15 games

### Q4: what are the top 5 games of each region by Number of Sales?

Lets look at top games of each region.

```python
fig, ax = plt.subplots(2, 2, figsize=(30, 20))
NA=vg_df.groupby(['Name'])[['NA_Sales']].sum()
EU=vg_df.groupby(['Name'])[['EU_Sales']].sum()
Japan=vg_df.groupby(['Name'])[['JP_Sales']].sum()
other=vg_df.groupby(['Name'])[['Other_Sales']].sum()
ax_NA = ax[0, 0]
NA.sort_values("NA_Sales", ascending=False).head(5).plot(kind='barh', ax=ax_NA,)
ax_NA.set_xlabel("Millions")
ax_NA.set_ylabel("Game")
ax_NA.set_title("Top 5 NA Sales")


ax_EU = ax[0, 1]
EU.sort_values("EU_Sales", ascending=False).head(5).plot(kind='barh', ax=ax_EU)
ax_EU.set_xlabel("Millions")
ax_EU.set_ylabel("Game")
ax_EU.set_title("Top 5 EU Sales")

ax_japan = ax[1, 0]
Japan.sort_values("JP_Sales", ascending=False).head(5).plot(kind='barh', ax=ax_japan)
ax_japan.set_xlabel("Millions")
ax_japan.set_ylabel("Game")
ax_japan.set_title("Top 5 Japan Sales")

ax_other = ax[1, 1]
other.sort_values("Other_Sales", ascending=False).head(5).plot(kind='barh', ax=ax_other)
ax_other.set_xlabel("Millions")
ax_other.set_ylabel("Game")
ax_other.set_title("Top 5 Other Sales")

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```
